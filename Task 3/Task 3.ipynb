# Preprocessing

```python
#Convert labels from strings to numeric values
from sklearn.preprocessing import LabelEncoder
#CarName	fueltype	aspiration	doornumber	carbody	drivewheel	enginelocation fuelsystem cylindernumber 
le =  LabelEncoder()
df["CarName"] = le.fit_transform(df["CarName"])
df["fueltype"] = le.fit_transform(df["fueltype"])
df["aspiration"] = le.fit_transform(df["aspiration"])
df["doornumber"] = le.fit_transform(df["doornumber"])
df["carbody"] = le.fit_transform(df["carbody"])
df["drivewheel"] = le.fit_transform(df["drivewheel"])
df["enginelocation"] = le.fit_transform(df["enginelocation"])
df["fuelsystem"] = le.fit_transform(df["fuelsystem"])
df["fuelsystem"] = le.fit_transform(df["fuelsystem"])
df["cylindernumber"] = le.fit_transform(df["cylindernumber"])
df["enginetype"] = le.fit_transform(df["enginetype"])




df.head()
```

# Getting the model ready

```python
from sklearn.model_selection import train_test_split

labels = df["price"]
features = df.drop(columns=["price"])
#split data
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=.3)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(x_train)
scaler.fit_transform(x_test)
```

# Training the model 

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
print("Accuracy of the model: ",model.score(x_test,y_test)*100,"%")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("Accuracy of the model: ",model.score(x_test,y_test)*100,"%")
```
