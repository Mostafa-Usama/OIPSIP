# Importing modules 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Information about the dataset

```python
df = pd.read_csv("Unemployment in India.csv")
df.head(10)
```

```python
# we have some Null values 
df.isnull().sum()
```

```python
# 28 null value for each column, these are small records and dropping these record is good option
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

# Data Vizualization

```python
area = []
# Caluclate mean of  Estimated Unemployment Rate (%) for each area
for i in df.Area.unique():
    area.append(df.loc[df['Area'] == i, ' Estimated Unemployment Rate (%)'].mean())
    

unemploy = df[" Estimated Unemployment Rate (%)"]
plt.title("Estimated Unemployment Rate for each area")

plt.pie(area,labels=["Urban","Rural"],autopct="%0.2f",wedgeprops={'edgecolor':"black"},shadow=True,explode=[0,.1])

plt.legend()
# We notice that Rural areas have more average unemployment rate (%)
```

```python
region = []
# Caluclate mean of  Estimated Unemployment Rate (%) for each area
for i in df.Region.unique():
    region.append(df.loc[df.Region == i, ' Estimated Unemployment Rate (%)'].mean())
    

unemploy = list(df.Region.unique())

plt.title("Estimated Unemployment Rate for each area")
plt.ylabel("Area")
plt.xlabel("Average Unemployment Rate (%)")
plt.barh(unemploy,region,edgecolor="black",color="#08ABCD")

# average unemployment rate (%) per region
# Tripura has the highest average unemployment rate while Meghalaya has the lowest
```

```python
region = []
# Caluclate mean of Estimated employed for each area
for i in df.Region.unique():
    region.append(df.loc[df['Region'] == i, ' Estimated Employed'].mean())
    

unemploy = list(df.Region.unique())
plt.title("Estimated Employed for each area")
plt.ylabel("Area")
plt.xlabel("Estimated Employed in millions")
plt.barh(unemploy,region,edgecolor="black",color="#8BDB0D")
```

```python
region = []
# Caluclate mean of  Estimated Labour Participation Rate (%) for each area
for i in df.Region.unique():
    region.append(df.loc[df['Region'] == i, ' Estimated Labour Participation Rate (%)'].mean())
    

unemploy = list(df.Region.unique())
plt.title("Average Labour Participation Rate for each area")
plt.ylabel("Area")
plt.xlabel("Average Labour Participation Rate(%)")
plt.barh(unemploy,region,edgecolor="black",color="#5EEBB0")
# Tripura, Telangana and Meghalaya have high Labour Participation rates
```

```python
plt.scatter(x=df[' Estimated Unemployment Rate (%)'], y=df[' Date'])
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Date")
plt.title("Estimated Unemployment Rate (%) over time")
# we notice that starting from 2020, Estimated Unemployment Rate increased duo to COVID-19
```
